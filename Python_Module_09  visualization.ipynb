# Python Module 9 - Visualization with Matplotlib

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

## Basic line plot with lists

```python
a=[1, 2, 3, 4]   # the simplest plot possible
b=[2, 4, 8, 16]
plt.plot(a,b)
plt.show()
```

```python
plt.title("This is a Title")     # add title
plt.xlabel("X axis")             # x axis label
plt.ylabel("Y axis")             # y axis label
plt.plot(a,b)
plt.show()
```

## Line plots with NumPy 

```python
x = ('Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat') 
y = np.array([25, 23, 26, 23, 33, 25, 29])           # temperatures 
z = y * 1.8+32
plt.plot(x, y, color='green')
plt.plot(x, z)
plt.title("Centigrade and Fahrenheit")
plt.xlabel("Day")
plt.ylabel("Temp")
plt.legend(["Degrees C", "Degrees F"])
plt.show()
```

## Scatter plot with NumPy data 

```python
x = np.linspace(0, 10, 11)      #lnspace fills in values between 0 and 10
x
```

```python
y = x+ np.sin(x)              #simple function to give us some nonlinear points
y
```

```python
chart=plt.scatter(x,y)         #straight scatter plot, no labels or titles 
plt.show()
```

```python
plt.title("Shoes size v. Hat size", fontsize=50)        #you can adjust the size of the font 
plt.xlabel("Shoe size",fontsize=20)
plt.ylabel("Hat size", fontsize=20)
chart=plt.scatter(x,y)
plt.show()
```

## Bar charts (vertical and horizontal)

```python
x = ['small', 'medium', 'large']        #x-axis for bar charts is discrete (categorical) data, not continuous
y = [61, 70, 74]
plt.bar(x, y)
plt.title("How High")
plt.xlabel("size")
plt.ylabel("height")
plt.legend(["blue bars"])
plt.show()
```

```python
plt.barh(x, y)
plt.show()
```

# Matplotlib with pandas dataframes

```python
df = pd.read_csv('CountriesWorld.csv')
```

```python
df.columns
df.head(3)
```

```python
df.plot.scatter(x='GDP ($ per capita)',y='Phones (per 1000)')
plt.title("GDP vs phone", fontsize = 20)
plt.show()
```

```python
df.hist(column='GDP ($ per capita)', bins=50)
plt.show()
```

```python
df.plot(x='Country', y='GDP ($ per capita)', kind = 'bar')
```

## Bar chart for a pivot table 

```python
pivot_df = pd.pivot_table(df,index='Region', values='GDP ($ per capita)', aggfunc='mean')
```

```python
pivot_df
```

```python
pivot_df.plot( kind = 'bar')
plt.ylabel('$ per capita')
plt.title('GDP per capita by region')
plt.legend()
plt.show()
```

```python
pivot_df.plot.pie(y='GDP ($ per capita)', figsize=(5, 8), legend=0)
```

## Box plots with dataframes

```python
df.boxplot(by='Region', column='GDP ($ per capita)', rot=90)
plt.title('')
plt.ylabel('GDP $ per capita')
```

## Line plots with dataframes

```python
dow_df = pd.read_csv('DJIA.csv', parse_dates=True)
```

```python
dow_df.info()
```

```python
dow_df.head(3)
```

```python
dow_df = dow_df.dropna(axis='columns', how ='all')
```

```python
dow_df.head(3)
```

```python
dow_df = dow_df.drop(columns='Unnamed: 8' )
dow_df.head(3)
```

```python
dow_df['Date']=pd.to_datetime(dow_df['Date'])
```

```python
dow_df.plot(x='Date', y='Close', color='green')
```

## Create a rolling average to smooth the line

```python
#
plt.plot(dow_df['Date'],dow_df['Close'],label= 'Close')
plt.plot(dow_df['Date'],dow_df['Close'].rolling(365).mean(),label= 'Dow 365 average')
plt.legend(loc='best')
plt.title('DJIA Close and Moving Averages')
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 5)
# Divide the figure into a 1x2 grid, and give me the first section
ax1 = fig.add_subplot(121)
plt.plot(dow_df['Date'],dow_df['Close'],label= 'DJIA Close')
plt.legend(loc='best')
# Divide the figure into a 1x2 grid, and give me the second section
ax2 = fig.add_subplot(122)
plt.plot(dow_df['Date'],dow_df['Close'].rolling(365).mean(),label= 'Dow 365 average')
plt.legend(loc='best')
```

## Segementing data for comparison

```python
country_df = pd.read_csv('Population.csv')
```

```python
country_df
```

## Compare the population growth in the US, India, and China

```python
us = country_df[country_df.country == 'United States']
```

```python
china = country_df[country_df.country == 'China']
```

```python
india = country_df[country_df.country == 'India']
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.plot(india.year, india.population / 10**6)

plt.legend(['United States', 'China', 'India'])
plt.xlabel('year')
plt.ylabel('population (millions)')
plt.show()
```
